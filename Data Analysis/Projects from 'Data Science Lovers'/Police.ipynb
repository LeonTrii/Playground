# Police Dataset

The dataset and questions are from: https://youtu.be/GyUbo45mVSE

```python
import pandas as pd
data = pd.read_csv('Police.csv')
data.index = data['stop_date']
data.drop(columns = 'stop_date', inplace = True)
```

## Q. 2) Based on Filtering + Value Counts - For Speeding, were men or women stopped more often? 

Men were stopped more often for speeding. Men had 25517 instances while women had 11686 instances for speeding.

```python
speeding = data[data['violation'] == 'Speeding']
speeding[speeding['driver_gender'] == 'M']
```

```python
speeding[speeding['driver_gender'] == 'F']
```

```python
data.groupby('driver_gender').search_conducted.sum()
```

## Q. 3) Does which gender gets searched more during a stop?

Males get searched more during a stop.

```python
searched = data[data['search_conducted'] == True]
searched[searched['driver_gender'] == 'M']
```

```python
searched[searched['driver_gender'] == 'F']
```

```python
data.groupby('driver_gender').violation.value_counts()
```
