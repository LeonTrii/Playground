```python
import pandas as pd;
import seaborn as sns;
pd.options.mode.chained_assignment = None
data = pd.read_csv('Cars.csv');
data.index = data['Model']
data.drop(columns = ['Model'], inplace = True)
data
```

## Q. 1) Data Cleaning - Find all Null Values in the dataset. If there are any Null Values in any columns, fill it with the mean of that column.

```python
data.isnull().sum()
```

```python
data['Cylinders'].fillna(data['Cylinders'].mean(), inplace = True)
data[['Cylinders']]
```

## Q. 2) Based on Value Counts - Check the different types of make in the dataset. List the occurances of each make.

```python
data['Make'].unique()
```

```python
data['Make'].value_counts()
```

## Q. 3) Filtering - Show the records in which the origin is Asia or Europe.

```python
asia_europe = data[(data['Origin'] == 'Asia') | (data['Origin'] == 'Europe')]
asia_europe
```

## Q. 4) Removing unwanted records - Remove the records (rows) in which the weight is above 4000.

```python
data = data[data['Weight'] <= 4000]
data
```

## Q. 5) Applying a function on a column - Increase the values of the, 'MPG_City' column by 3.

```python
data['MPG_City'] += 3
data
```
